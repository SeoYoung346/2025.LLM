```python
from openai import OpenAI
```

```python
client = OpenAI(api_key=OPENAI_API_KEY)
myfile=client.files.create(
  file=open("/content/drive/MyDrive/지능형소프트웨어/finetune_data.jsonl", "rb"),
  purpose="fine-tune"
)
print(myfile.id)
```

```python
myjob = client.fine_tuning.jobs.create(
    training_file=myfile.id,
    model="gpt-3.5-turbo"
)
print(myjob.id)
```

```python
print(myjob.status)
```

```python
job = client.fine_tuning.jobs.retrieve(myjob.id)

if job.status == 'succeeded':
    fine_tuned_model_id = job.fine_tuned_model

    completion = client.chat.completions.create(
        model=fine_tuned_model_id, # Use the correct model ID
        messages=[
            {"role": "system", "content": "네모바지 스폰지밥의 뚱이 캐릭터의 말투를 사용하여 답장하시오. 뚱이는 엉뚱하며 친구를 아끼는 캐릭터이다."},
            {"role": "user", "content": "안녕, 뚱아!"}
        ]
    )
    print(completion.choices[0].message.content)
else:
    print("Fine-tuning job is still in progress. Please wait and try again later.")
```
